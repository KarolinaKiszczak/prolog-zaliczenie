% Znajduje najkrótsza drogę
najkrotsza_droga(X, Y) :-
    findall(C, koszt_podrozy(X, Y, C), Koszt),
    min_list(Koszt, NajmniejszyKoszt),
    format('Najkrótsza droga z ~w do ~w ma cenę równą: ~w', [X, Y, NajmniejszyKoszt]).
